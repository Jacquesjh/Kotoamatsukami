# Importing Libs

```python
import pickle

import cv2

from kotoamatsukami import HandDetector
```

# Set the Distance!

```python
distance = "50cm"
```

# Gestures Info Handling

```python
gestures_names = {

    0: "click",
    1: "closed",
    2: "down",
    3: "mouse_tracking",
    4: "negative_closed",
    5: "negative_mouse_tracking",
    6: "negative_side",
    7: "negative_up",
    8: "side",
    9: "up"

}
```

# Aux functions

```python
def save_features(distance: str, gest_dict: dict) -> None:
    for gesture_name, features in gest_dict.items():
        with open(f"data/streamed/{distance}/{gesture_name}", "wb") as file:
            pickle.dump(features, file)
```

# Main

```python
gestures_features = {name: list() for name in list(gestures_names.values())}

num_frames_per_gesture = 1500
```

```python
stream   = cv2.VideoCapture(0)
detector = HandDetector(max_hands = 1, detection_con = 0.85, normalize = False)

for gesture_name in list(gestures_names.values()):
    gesture_image = cv2.imread(f"gestures/{gesture_name}.png", 0)
    
    OK = False
    
    while True:
        print(len(gestures_features.get(gesture_name)))

        if len(gestures_features.get(gesture_name)) > num_frames_per_gesture:
            break

        if cv2.waitKey(1) == ord("x"):
            break

        if cv2.waitKey(1) == ord("t"):
            OK = True
       
        _grabbed, raw_frame = stream.read()

        hands, image = detector.find_hands(image=raw_frame, draw_box=False, draw_marks=False)

        cv2.imshow("Video", image)
        cv2.imshow("Gesture", gesture_image)

        if len(hands) != 0:
            features = hands[0]["landmarks"]

        if OK:
            gestures_features[gesture_name].append(features)

# save_features(distance = distance, gest_dict = gestures_features)

stream.release()
cv2.destroyAllWindows()
```
